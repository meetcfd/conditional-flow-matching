```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x,s=1,sig=0.1,sig_min=1e-2: (1. - x)*np.exp(-s*x)*sig + sig_min
f2 = lambda x,s=1,sig=0.1,sig_min=1e-2: -sig*np.exp(-s*x)*(s*(1-x) + 1)/(f(x,s, sig, sig_min))  # outse
f3 = lambda t: (1 - 2 * t) / (2 * t * (1 - t) + 1e-8) # schrodinger bridge proposal
f4 = lambda t, sig: -(1 - sig)/(1 - (1 - sig)*t) # Target Conditional Flow Matching
x = np.linspace(0,1,100)
```

```python
s=5
```

```python
plt.plot(f(x,s, 1e-1, 1e-2))
```

```python
plt.plot(f2(x,s,1e-2, 1e-2))
```

```python
plt.plot(f3(x))
```

```python
plt.plot(f4(x, 1e-1))
```
