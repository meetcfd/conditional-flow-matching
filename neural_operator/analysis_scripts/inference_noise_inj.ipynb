```python
exps_path ="/home/meet/FlowMatchingTests/conditional-flow-matching/neural_operator/exps/ks_high_res_noise_inj/cs_exps"
```

```python
exps = {i:128 for i in range(1,6)}
exp=3
for iter in [19900]:
    ot_cfm_model = std_MLP([[256, 256, exps[exp]], [exps[exp], exps[exp], exps[exp]], [exps[exp], exps[exp], exps[exp]], [exps[exp], exps[exp], exps[exp]] ,[exps[exp], 256, 256]],
                        torch.nn.ReLU())
    state = torch.load(f"{exps_path}/exp_{exp}/saved_state/checkpoint_{iter}.pth")
    ot_cfm_model.load_state_dict(state["model_state_dict"])
    ot_cfm_model.eval()

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    ot_cfm_model.to(device);
    test_avg_corr = []
    for num in range(800,993):
        with torch.no_grad():
            out_lst = [X[num:num+1,0]]
            init = torch.from_numpy(X[num:num+1,0]).float().to(device)
            for i in range(512):
                inp = init if i == 0 else torch.from_numpy(out).float().to(device)
                out = ot_cfm_model(inp).detach().cpu().numpy()
                out_lst.append(out)
        test_avg_corr.append((pearsonr(np.concatenate(out_lst)[:512], X[num, :512], axis=-1)[0] >= 0.8).sum())
    hist, bin_edges = np.histogram(np.array(test_avg_corr), bins=10, density=True)
    plt.plot((bin_edges[:-1] + bin_edges[1:])/2, hist, alpha=0.5, label=iter);
plt.legend()
plt.show()
```
